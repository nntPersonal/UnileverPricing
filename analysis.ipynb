```python
import pandas as pd
```

```python
# Read the data from the CSV file Product Pricing
price_df = pd.read_csv('pricing.csv')

# drop Ngân Sách column
price_df = price_df.drop(columns=['Ngân Sách'])

price_df
```

```python
# standardize column Brand to upper case
price_df['Brand'] = price_df['Brand'].str.upper()
```

```python
# loop through each row in the DataFrame, if there is any data in the column After CK 1, convert it to float. Skip if there's null
for index, row in price_df.iterrows():
    try:
        price_df.at[index, 'After CK 1'] = float(row['After CK 1'])
        price_df.at[index, 'After CK 2'] = float(row['After CK 2'])
    except:
        pass
```

```python
def filter_rows_by_keywords(df, column_name, input_string):
    # Separate the input string into individual keywords
    keywords = input_string.split()
    
    # Function to check if all keywords are in the cell
    def contains_all_keywords(cell_value):
        return all(keyword.lower() in cell_value.lower() for keyword in keywords)
    
    # Apply the function to filter the dataframe
    filtered_df = df[df[column_name].apply(contains_all_keywords)]
    
    return filtered_df
```
